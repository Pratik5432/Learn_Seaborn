### 1.Relational Plots
- to see the statistical relation between 2 or more variables
- Bivariate Analysis

Plots under Relational Plots
- LinePlot
- ScatterPlot

```python
tips = sns.load_dataset('tips')
tips
```

```python
#scatter plot -> axes level function
sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# relplot -> figure level function
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# Facet Plot (relplot) : Works only with figure level function
sns.relplot(tips,x='total_bill',y='tip',kind='scatter',row='smoker',col='sex',style='time',size='size')
```

```python
gap = px.data.gapminder()
temp = gap[gap['country'] == "India"]
temp
```

```python
# line plot -> axes level function
sns.lineplot(temp,x='year',y='lifeExp')
```

```python
# relplot -> figure level function
sns.relplot(temp,x='year',y='lifeExp',kind='line')
```

```python
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.lineplot(temp_df,x='year',y='lifeExp',hue='country',style='continent',size ='continent')
```

```python
sns.relplot(temp_df,x='year',y='lifeExp',hue='country',kind='line',style='continent',size ='continent')
```

```python
# Facet plot
sns.relplot(temp_df,x='year',y='lifeExp',row='country',kind='line',style='continent',size ='continent')
```

### 2.Distribution Plots
- used to find out the distribution
- Range of observation
- Univariate Analysis
- Central Tendency
- Is the data bimodel?
- Are there outliers?

Plots under Relational Plots
- HistPlot
- RugPlot
- kdeplot

```python
#HistPlot
sns.histplot(tips,x='total_bill',bins=20,hue='sex')
```

```python
sns.displot(tips,x='total_bill',kind='hist',bins=20,hue='sex',element='step')
```

```python
sns.histplot(tips,x='day')
```

```python
# Facet Function
sns.displot(tips,x='total_bill',kind='hist',bins=20,col='sex')
```

```python
# kdeplot : Rather than using discrete bins, a kde plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(tips,x='total_bill')
```

```python
sns.displot(tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
#Rugplot
# Plot marginal distributions by drawing ticks along the x and y axes
# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(tips,x='total_bill')
sns.rugplot(tips,x='total_bill')
```

```python
# Bivariate histplot
sns.histplot(tips,x='total_bill',y='tip')
```

```python
sns.displot(tips,x='total_bill',y='tip',kind='hist')
```

```python
# Bivariate kdeplot
sns.kdeplot(tips,x='total_bill',y='tip')
```

```python
sns.displot(tips,x='total_bill',y='tip',kind='kde')
```

### 3.Matrix Plots

Plots under Relational Plots
- Heatmap
- Clustermap

```python
temp_var = gap.pivot(index='country',columns='year',values='lifeExp')
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(temp_var)
```

```python
temporary = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(temporary, annot=True, linewidth=0.5, cmap='jet')
```

```python
iris = px.data.iris()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
