```python
import seaborn as sns
import plotly.express as px
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
#Strip Plot
sns.stripplot(tips,x='day',y='total_bill',hue='sex')
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='strip',hue='sex')
```

```python
#Swarm Plot
sns.swarmplot(tips,x='day',y='total_bill')
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='swarm',hue='sex')
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
sns.boxplot(tips,x='sex',y='total_bill')
```

```python
sns.boxplot(tips,x='day',y='total_bill',hue='sex')
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='box')
```

```python
sns.boxplot(tips,y='total_bill')
```

```python
#Violin Plot
sns.violinplot(tips,x='day',y='total_bill',hue='sex',split=True)
```

```python
sns.catplot(tips,x='day',y='total_bill',kind='violin')
```

```python
#Bar Plot
import numpy as np
sns.barplot(tips,x='sex',y='total_bill',estimator=np.std,hue='smoker')
```

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars.

```python
sns.catplot(tips,x='sex',y='total_bill',kind='bar')
```

```python
# Point Plot
sns.pointplot(tips,x='sex',y='total_bill',estimator=np.min,hue='smoker')
```

```python
sns.catplot(tips,x='sex',y='total_bill',kind='point')
```

```python
# count Plot
sns.countplot(tips,x='sex',hue='day')
```

```python
sns.catplot(tips,x='sex',kind='count')
```

### 4.Regression plot
- Regplot (axes level function)
- Implot  (figure level function)

In the simplest invocation, both functions draw a scatterplot of 2 variables x and y, & then fit the regression model y ~ x, and plot the resulting regression line & a 95% confidence interval for that regression.

```python
sns.regplot(tips,x='total_bill',y='tip')  # hue parameter is not present
```

```python
sns.lmplot(tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot
sns.residplot(tips,x='total_bill',y='tip')
```

### Multi Grid Plots

```python
g = sns.FacetGrid(tips,col='day',row='time',hue='smoker')
g.map(sns.violinplot,'sex','total_bill')
```

```python
g = sns.FacetGrid(tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')
```

### Plotting Pairwise Relationship (PairGrid vs PairPlot)

```python
iris
```

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(iris,hue='species')

g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.violinplot)
g.map_offdiag(sns.histplot)
```

```python
g = sns.PairGrid(iris,hue='species')
g.map_diag(sns.violinplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.histplot)
```

```python
g = sns.PairGrid(iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.violinplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.histplot)
```
